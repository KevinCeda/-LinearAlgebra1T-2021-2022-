##Determinant
A determinant is a scalar value derived from a square matrix. The determinant is a fundamental and important value used in matrix algebra. Although it will not be evident in this laboratory on how it can be used practically, but it will be reatly used in future lessons.

```python
A = np.array([
    [1,4],
    [0,3]
])
np.linalg.det(A)
```

##Inverse
The inverse of a matrix is another fundamental operation in matrix algebra. Determining the inverse of a matrix let us determine if its solvability and its characteristic as a system of linear equation — we'll expand on this in the nect module. Another use of the inverse matrix is solving the problem of divisibility between matrices. Although element-wise division exists but dividing the entire concept of matrices does not exists. Inverse matrices provides a related operation that could have the same concept of "dividing" matrices.

```python
M = np.array([
   [1,7],
   [-3,5]           
])

np.array(M @ np.linalg.inv(M), dtype=int)
```

```python
P = np.array([
              [6, 9, 0],
              [4, 2, -1],
              [3, 6, 7]
])
Q = np.linalg.inv(P)
Q
```

```python
P @ Q
```

A matrix's data can be integers, strings, equations, symbols, and so on. The matrix is one of the most significant data structures for mathematical and scientific operations. In this laboratory activity, One of the most significant functions in matrix multiplication is the transpose() function. The dimension of the specified array is permuted or reserved by this function, which returns the changed array. The numpy.transpose() method transforms row elements into column elements and column elements into row elements. Taking the sum of a list comprehension and performing element-wise multiplication is one approach to calculate the dot product in Python. Alternatively, the np.dot() method can be used. The Determinant of a square matrix is a particular integer that may be derived from a square matrix. The numpy.linalg.det() function in Numpy allows us to determine the determinant of a square matrix. The inverse of a matrix may be calculated quickly and easily in Python.
